```python
import pandas as pd
```

```python
# df is daily data downloaded from yahoo
df  = pd.read_csv("AAPL.csv") 
# df1 is weekly data downloaded from yaho
df1 = pd.read_csv("AAPL_W.csv")
```

```python
def getExpectedReturn(df, price_col_name, annualised=True, annualise_method='sophisticated',  frequency='daily'):
    """
    Returns the expected return of a stock given price data.
    """

    # Calculate returns of prices - daily
    if frequency:
        if frequency == 'daily':
            data = df
            returns = data[price_col_name].pct_change(1)
            expected_return = returns.mean()
        elif frequency == 'weekly': 
            data = df1
            returns = data[price_col_name].pct_change(1)
            expected_return = returns.mean()
            
        return expected_return
    else:
        pass

    if annualised:
        if  frequency == 'daily' and  annualise_method == 'sophisticated':
            expected_return_annual = ((1 + expected_return) ** 250) - 1
        elif frequency == 'daily' and annualise_method == 'crude':
            # Crude method
            expected_return_annual = expected_return * 250
            
        elif frequency == 'weekly' and annualise_method == 'sophisticated':
            expected_return_annual = ((1 + expected_return) ** 52) - 1
        elif frequency == 'weekly' and annualise_method == 'crude':
            expected_return_annual = expected_return * 52

        return expected_return_annual
        

    else:
        return expected_return
```

```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AAPL.csv")
df = df[['Date', 'Adj Close']] 

df.rename(
    columns={'Date' : 'date', 'Adj Close' : 'price_t'},
    inplace=True)  

df['returns'] = df['price_t'].pct_change(1) 
```

```python
df['returns']
```

```python
var_aapl = np.var(df['returns'], ddof=1)
sd_aapl = np.sqrt(var_aapl)
```
